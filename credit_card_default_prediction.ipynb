```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

**--> Reading and analysing the dataset**

```python
credit_card_data=pd.read_csv('/content/CreditCardDefault.csv')
```

```python
# information about dataset
credit_card_data.info()
```

```python
# check missing values 
credit_card_data.isnull().sum()
```

```python
# check distribution of legit and fraud transaction
credit_card_data.Class.value_counts()
```

**Dataset is highly unbalanced as legit transactions >>> fraud/default transaction** 

0 --> legit transaction

1 --> fraud transaction

```python
# separating data into legit and fraud dataset

legit=credit_card_data[credit_card_data.Class ==0]
fraud=credit_card_data[credit_card_data.Class ==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# analysing legit and fraud data

legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the mean values for both the transactions

credit_card_data.groupby('Class').mean()
```

**Build another sample dataset containing similar distribution of legit and fraud transactions**

fraud transactions --> 492 , 
 legit transactions --> 284315


**Take 492 transactions randomly from legit transactions and then combine it with fraud transactions , so that distribution become balanced**

```python
legit_sample=legit.sample(n=492)
```

```python
# concatenating fraud and legit_sample datasets
new_dataset=pd.concat([legit_sample,fraud]) # our ml model will be trained on this dataset
new_dataset.shape
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset.Class.value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
new_dataset.columns
```

```python
# separating features and targets
X=new_dataset[['Time', 'V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9', 'V10',
       'V11', 'V12', 'V13', 'V14', 'V15', 'V16', 'V17', 'V18', 'V19', 'V20',
       'V21', 'V22', 'V23', 'V24', 'V25', 'V26', 'V27', 'V28', 'Amount']]
y=new_dataset['Class']
```

```python
print(X.shape)
print(y.shape)
```

```python
# train-test split
X_train,X_test,y_train,y_test=train_test_split(X,y,train_size=0.8,stratify=y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

**Model training**

```python
# train machine learning model with training data
model=LogisticRegression()
model.fit(X_train,y_train)
```

**Evaluation of model**

```python
# calculating accuracy on training data

X_train_prediction=model.predict(X_train)
training_data_accuracy= accuracy_score(X_train_prediction,y_train)
```

```python
print('Accuracy on training data : ',training_data_accuracy)
```

```python
# calculating accuracy on test data

X_test_prediction=model.predict(X_test)
test_data_accuracy= accuracy_score(X_test_prediction,y_test)
```

```python
print('Accuracy on test data : ',test_data_accuracy)
```
